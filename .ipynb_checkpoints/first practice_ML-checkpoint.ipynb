# Student Dataset

```python
import pandas as pd 
data={
    "Study hours":[2,4,6,8,10],
    "Attendance":[50,60,70,80,90],
    "Marks":[50,60,75,90,98]
}

df=pd.DataFrame(data)
df
```

```python
#Features(x)
x=df[["Study hours","Attendance"]]
x
```

```python
#Target(Y)
y=df[["Marks"]]
print(y)
```

# House Price Dataset

```python
import pandas as pd
data = {
    "Area": [1000, 1200, 1500, 1800],
    "Bedrooms": [2, 2, 3, 4],
    "Price": [200000, 250000, 320000, 450000]
}

df=pd.DataFrame(data)
df
```

```python
# Features (X)
x=df[["Area","Bedrooms"]]
print(x)
```

```python
#Taget (Y)
y=df[["Price"]]
print(y)
```

# Employee Salary Dataset

```python
import pandas as pd 
data = {
    "Experience": [1, 2, 3, 4, 5],
    "Age": [22, 24, 26, 28, 30],
    "Salary": [30000, 40000, 50000, 60000, 70000]
}

df=pd.DataFrame(data)
print(df)
```

```python
# Features (X)
x=df[["Experience","Age"]]
print(x)
```

```python
# Target(Y)
y=df[["Salary"]]
print(y)
```
